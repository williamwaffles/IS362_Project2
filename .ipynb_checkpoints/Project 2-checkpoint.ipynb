## Untidy Dataset 1: 
### Load Car MPG .csv into Dataframe

#### Data from https://archive.ics.uci.edu/ml/datasets/Auto+MPG

```python
mpg_df = pd.read_csv('auto-mpg.data-original.csv')
mpg_df
```

### Reload MPG data into dataframe with headers

```python
car_df = pd.read_csv('auto-mpg.data-original.csv', names=['MPG','Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year', 'Car Name', 'Null1', 'Null2'])

car_df
```

### Remove columns with Null values

```python
car_df = car_df.drop(columns = ['Null1', 'Null2'])
car_df
```

### Remove 1.\t, 2.\t & 3.\t from Car Name column

```python

strip_car_name = car_df['Car Name'].str.strip('1.\t')

car_df['Car Name'] = strip_car_name

strip_car_name = car_df['Car Name'].str.strip('2.\t')

car_df['Car Name'] = strip_car_name

strip_car_name = car_df['Car Name'].str.strip('3.\t')

car_df['Car Name'] = strip_car_name

car_df
```

### MPG data for cars with 4 (or fewer) cylinders

```python
mpg_by_cylinders1 = car_df.loc[car_df['Cylinders'] <= 4]
  
print('\nResult dataframe :\n', mpg_by_cylinders1)
```

### MPG data for cards with more than 4 cylinders

```python
mpg_by_cylinders2 = car_df.loc[car_df['Cylinders'] > 4]
  
print('\nResult dataframe :\n', mpg_by_cylinders2)
```

## Untidy Dataset 2:
### Load Caffeine content .csv into Dataframe
### .csv posted by Sarah Bill

```python
sb_df = pd.read_csv('starbucks_caffeine.csv')
sb_df
```

### Remove Null/NaN values

```python
sb_df.fillna('No data', inplace=True)
sb_df
```

### Replace '-' with 'No data' string

```python
sb_df = sb_df.replace('–', 'No data')
sb_df
```

### Caffeine content sorted by Short size, descending

```python
sb_df.sort_values(by='Short (8 fl oz)')
```

## Untidy Dataset 3:
### Data Scientist Salaries
### .csv posted by Daniel Collins 

```python
ds_df = pd.read_csv('ds_salaries_untidy.csv')
ds_df
```

### Remove null values

```python
ds_df.fillna('No data', inplace=True)
ds_df
```

### Find the job_title with the highest salary

```python
ds_df.loc[ds_df['salary_in_usd'].idxmax()]
```

### Find the job_title with the lowest salary

```python
ds_df.loc[ds_df['salary_in_usd'].idxmin()]
```

### Salary sorted by company_location and job_title

```python
ds_df.groupby(['company_location','job_title'])['salary_in_usd'].max()
```
