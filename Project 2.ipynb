### Import Pandas

```python
import pandas as pd
```

### Load Car MPG .csv into Dataframe

#### Data from https://archive.ics.uci.edu/ml/datasets/Auto+MPG

```python
mpg_df = pd.read_csv('auto-mpg.data-original.csv')
mpg_df
```

### Reload MPG data into dataframe with headers

```python
car_df = pd.read_csv('auto-mpg.data-original.csv', names=['MPG','Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year', 'Car Name', 'Null1', 'Null2'])

car_df
```

### Remove columns with Null values

```python
car_df = car_df.drop(columns = ['Null1', 'Null2'])
car_df
```

### Remove 1.\t, 2.\t & 3.\t from Car Name column

```python

strip_car_name = car_df['Car Name'].str.strip('1.\t')

car_df['Car Name'] = strip_car_name

strip_car_name = car_df['Car Name'].str.strip('2.\t')

car_df['Car Name'] = strip_car_name

strip_car_name = car_df['Car Name'].str.strip('3.\t')

car_df['Car Name'] = strip_car_name

car_df
```

### MPG data for cars with 4 

```python
mpg_by_cylinders1 = car_df.loc[car_df['Cylinders'] <= 4]
  
print('\nResult dataframe :\n', mpg_by_cylinders1)
```

```python
mpg_by_cylinders2 = car_df.loc[car_df['Cylinders'] > 4]
  
print('\nResult dataframe :\n', mpg_by_cylinders2)
```
